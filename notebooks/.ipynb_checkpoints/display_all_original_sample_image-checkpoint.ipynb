### - JPG Format

#### - Fair Image

```python
img_fair_attraction = imageio.imread('../images/jpg/fair_attraction.jpg')

plt.imshow(img_fair_attraction)
plt.title('Fair Attraction')
plt.show()

print(f'Height        : {img_fair_attraction.shape[0]} pixels')
print(f'Width         : {img_fair_attraction.shape[1]} pixels')
print(f'Color Channel : {img_fair_attraction.shape[2]} (RGB))')
```
